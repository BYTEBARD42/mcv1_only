```python
import pandas as pd
import joblib
from mcv1_forecast.core import engineer_features, recursive_forecast, TARGET
import warnings
warnings.filterwarnings("ignore")

df_raw = pd.read_csv("vaccine_data.csv")
df_future = pd.read_csv("future_demographics.csv")
```

```python
# Load model and columns
saved_data = joblib.load('mcv1_model.pkl')
model = saved_data['model']
feature_cols = saved_data['feature_cols']
dummy_cols = saved_data['dummy_cols']
```

```python
# Engineer historical features needed for lags
df_full, _, _ = engineer_features(df_raw, TARGET)
```

```python
# Forecast 2025-2030 using future demographics
split_year = 2025
forecast_results = recursive_forecast(df_full, model, feature_cols, dummy_cols, split_year, future_demo_df=df_future)

forecast_results
```

```python
# Save forecasts
forecast_results.to_csv("forecast_results.csv", index=False)
print("Forecasts saved to forecast_results.csv")
```
